# Introduction

## Diabetes is a serious disease affecting millions of people across the Globe. The high blood sugar level is the primary cause mostly seen in this disease.
## This dataset explains the various health parameters and helps in diagnostically predict whether a patient has diabetes or not. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

### variables used in this dataset are:

### Columns of the dataset

**Pregnancies**: Number of times pregnant

**Glucose**: Plasma glucose concentration 2 hours in an oral glucose tolerance test

**Blood Pressure**: Diastolic blood pressure (mm Hg)

**Skin Thickness**: Triceps skin fold thickness (mm)

**Insulin**: 2-Hour serum insulin (mu U/ml)

**BMI**: Body mass index (weight in kg/(height in m)^2) 

**Diabetes Pedigree Function**: indicates the function which scores likelihood of diabetes based on family history

**Age**: Age (years)

**Outcome**: Class variable (0 or 1) 0 mean non-diabetic and 1 means diabetic




# understanding the dataset

**Skin Thickness**:
A value used to estimate body fat estimation at all ages including the newborn period.

**insulin**:
insulin is a hormone created by your pancreas that controls the amount of glucose in your bloodstream at any given moment.

**BMI**:
Body mass index (BMI) is a simple index of weight-for-height that is commonly used to classify overweight and obesity in adults. It is defined as a person's weight in kilograms divided by the square of his height in meters (kg/m2).

BMI VALUES:
• Underweight (Below 18.5)

• Normal (18.5 - 24.9)

• Overweight (25.0 - 29.9)

• Obese (30.0 and Above)

**NORMALLY DISTRIBUTED**:   only Glucose , blood pressure , BMI , skin thickness graphs are normally distributed.

**SKEWED**:    Rest of the graphs i.e Pregnancies, , Insulin, DiabetesPedigreeFunction, Age are rightly skewed.

```python
#Q8. Please calculate the skewness value and divide variables into symmetrical, moderately skewed and highly skewed.


print("skewness for pregnancies is",diab["Pregnancies"].skew(),":  moderately skewed")    # .skew() function gives the skew value

print("skewness for BMI is",diab["BMI"].skew(),":  normally distributed") 

print("skewness for Glucose is",diab["Glucose"].skew(),":  normally distributed") 

print("skewness for BloodPressure is",diab["BloodPressure"].skew(),":  normally distributed") 

print("skewness for SkinThickness is",diab["SkinThickness"].skew(),":  normally distributed") 

print("skewness for DiabetesPedigreeFunction is",diab["DiabetesPedigreeFunction"].skew(),":  Highly skewed") 

print("skewness for Age is",diab["Age"].skew(),":  Highly skewed") 

print("skewness for Insulin is",diab["Insulin"].skew(),":  Highly skewed") 
```


## A skewness value greater than 1 or less than -1 indicates a highly skewed distribution. A value between 0.5 and 1 or -0.5 and -1 is moderately skewed. A value between -0.5 and 0.5 indicates that the distribution is normally distributed.
